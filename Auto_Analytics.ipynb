## 4.10.1 So let's identify the outlier using  the IQR method to check the data quality

```python

def detect_outliers_iqr(df1, columns):
    """
    Detects and prints the number of outliers in each column using the IQR method.

    Parameters:
        df1 (pd.DataFrame): The DataFrame containing the data.
        columns (list): List of numerical column names to analyze.

    Returns:
        dict: Dictionary with number of outliers per column.
    """
    outlier_counts = {}

    for col in columns:
        Q1 = df1[col].quantile(0.25)
        Q3 = df1[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        outliers = df1[(df1[col] < lower_bound) | (df1[col] > upper_bound)]
        count = outliers.shape[0]
        outlier_counts[col] = count
      

    return outlier_counts

cols_to_check = ['REPAIR_AGE', 'KM', 'REPORTING_COST', 'TOTALCOST', 'LBRCOST']

# Call the function
detect_outliers_iqr(df1, cols_to_check)
```


### So the IQR based outlier detection says that

* Most data is clean.

### 4 Generating tags/features from free text available 

```python
# Importing necessary libraries
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import nltk
# Download necessary resources
nltk.download('stopwords')
nltk.download('wordnet')

# assigning stopwords to the variable stop_words and creating an instance of WordNetLemmatizer
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def preprocess_text(text):
    tokens = text.split()
    # Remove stopwords and lemmatize tokens
    tokens = [lemmatizer.lemmatize(word.lower()) for word in tokens if word.lower() not in stop_words]
    return tokens

df1['correction_tokens'] = df1['CORRECTION_VERBATIM'].apply(preprocess_text)




```

```python
# In the following tags will be created based on the customer verbatim

def preprocess_text_customer(text):
    tokens = text.split()
    # Remove stopwords and lemmatize tokens
    tokens = [lemmatizer.lemmatize(word.lower()) for word in tokens if word.lower() not in stop_words]
    return tokens

df1['customer_tokens'] = df1['CUSTOMER_VERBATIM'].apply(preprocess_text_customer)


```
